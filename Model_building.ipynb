```python
x=final_rating['user_id'].value_counts() > 60
```

```python
dummy_final=final_rating[final_rating['user_id'].isin(x[x].index)]
```

```python
dummy_book_pivot=dummy_final.pivot_table(columns='user_id',index='title',values='rating')
```

```python
dummy_book_pivot.head(20)
```

```python
dummy_book_pivot.fillna(0,inplace=True)
```

```python
dummy_book_pivot.index[0]
```

```python
from scipy.sparse import csr_matrix
```

```python
book_sparse=csr_matrix(dummy_book_pivot)
```

```python
from sklearn.neighbors import NearestNeighbors
model=NearestNeighbors(algorithm='brute')
```

```python
model.fit(book_sparse)
```

```python
distances, suggestion = model.kneighbors(dummy_book_pivot.iloc[6,:].values.reshape(1,-1),n_neighbors=6)
```

```python
suggestion
```

```python
for i in range(len(suggestion)):
    print(dummy_book_pivot.index[suggestion[i]])
```

```python
books_name=dummy_book_pivot.index
```

```python
books_name
```

```python
def recommend_books(book_name):
    book_index=np.where(dummy_book_pivot.index==book_name)[0][0]
    distances, suggestion = model.kneighbors(dummy_book_pivot.iloc[6,:].values.reshape(1,-1),n_neighbors=6)
    for i in range(len(suggestion)):
        books=(dummy_book_pivot.index[suggestion[i]])
        for j in books:
            print(j)
```

```python
recommend_books('24 Hours')
```
